```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np



fonte =  "https://github.com/alura-cursos/imersao-dados-2-2020/blob/master/MICRODADOS_ENEM_2019_SAMPLE_43278.csv?raw=true"

dados = pd.read_csv(fonte)
```

```

dados["NU_NOTA_TOTAL"] =dados[provas].sum(axis = 1)
provas = ["NU_NOTA_CN", "NU_NOTA_CH", "NU_NOTA_MT", "NU_NOTA_LC", "NU_NOTA_REDACAO","NU_NOTA_TOTAL"]
dados_sem_notas_zeros = dados.query("NU_NOTA_TOTAL != 0")
plt.figure(figsize=(10,6))
sns.histplot(dados_sem_notas_zeros, x = "NU_NOTA_TOTAL")
```

**Desafio3: Filtrar as notas por seu estado ou cidade e rafazer as análises, verificando se são semelhantes ao geral**

```
correlacaoSP = dados_sem_notas_zeros.query("SG_UF_RESIDENCIA == 'SP'")[provas].corr()
maskSP = np.zeros_like(correlacaoSP)
maskSP[np.triu_indices_from(maskSP)] = True
plt.subplots(figsize=(10, 10))
sns.heatmap(correlacaoSP, cmap="Blues", center=0, annot=True, mask=maskSP)
```

```
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
#-----------------------------Geral----------------------------------
mask = np.zeros_like(correlacao)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(correlacao, cmap="Greens", center=0, annot=True, mask=mask)
plt.title("Geral")
#-----------------------------SP-------------------------------------
plt.subplot(2,2,2)
correlacaoSP = dados_sem_notas_zeros.query("SG_UF_RESIDENCIA == 'SP'")[provas].corr()
maskSP = np.zeros_like(correlacaoSP)
maskSP[np.triu_indices_from(maskSP)] = True
sns.heatmap(correlacaoSP, cmap="Blues", center=0, annot=True, mask=maskSP)
plt.title("São Paulo")
```
