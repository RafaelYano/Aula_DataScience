```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


fonte =  "https://github.com/alura-cursos/imersao-dados-2-2020/blob/master/MICRODADOS_ENEM_2019_SAMPLE_43278.csv?raw=true"

dados = pd.read_csv(fonte)
```

```
#verificar alunos de 13 anos feito no desafio da aula 1
dados[["NU_IDADE", "SG_UF_RESIDENCIA"]].value_counts().sort_index()
```

```
#Pessoas que realizaram aprova em outro estado
perc =( dados.query("SG_UF_RESIDENCIA != SG_UF_PROVA" ).value_counts().sum() / dados.value_counts().sum()) * 100
print(perc)
```

Desafio3: criar uma função para plotar os boxplots do seaborn

```
def my_BoxPlot(c, d):
  plt.figure(figsize=(10,8))
  sns.boxplot(x="Q006", y = c, data = d, order = renda)
  plt.title("gráfico da " + c+ " de acordo com a renda do aluno")
```

```
my_BoxPlot("NU_NOTA_CN", dados)
```

```
from matplotlib.lines import Line2D
plt.figure(figsize=(14,8))
legend_elements = [Line2D([0], [0], color='blue', lw=4 ),
Line2D([0], [0], color='orange', lw=4)]
sns.boxplot(x="Q025", y = "NU_NOTA_TOTAL", data = dados)
plt.title("Nota total dos alunos que possuem ou não internet")
plt.legend(handles = legend_elements, labels = ["Com internet", "Sem internet"] )
plt.show()
```
